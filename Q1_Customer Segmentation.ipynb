```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('customer_purchases.csv')
```

```python
df['customer_id'] = df['customer_id'].astype('str')
df['product_id'] = df['product_id'].astype('str')
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
```

```python
df.dtypes
```

```python
df2=df.groupby('customer_id')[['amount_spent']].sum()
```

```python
df2.head()
```

```python
def segment(x):
  if x < 100:
    return 'Low'
  elif x >= 100 and x < 500:
    return 'Medium'
  else:
    return 'High'

df2['Segment']=df2['amount_spent'].apply(segment)
df2.value_counts('Segment')
```

```python
df2.to_csv('customer_segments.csv')
```
